### 1. Оператори циклів в Python

Цикл - це конструкція мови програмування, що дозволяє багаторазово виконувати один і той самий блок коду. У Python існують два основні типи циклів:

#### 1.1 Цикл `while`
Цикл `while` виконує блок коду, **доки задана умова є істинною** (True). Використовується, коли кількість ітерацій заздалегідь невідома.

**Синтаксис:**
```python
while умова:
    # блок коду, що повторюється
```

````python
# Приклад 1: Простий цикл while
print("Цикл while:")
counter = 0
while counter < 5:
    print(counter)
    counter += 1
````

#### 1.2 Управління циклом while

У циклах можна використовувати оператори керування:
- `break` - негайно завершує виконання циклу
- `continue` - пропускає поточну ітерацію та переходить до наступної
- `else` - виконується після завершення циклу, якщо не було викликано `break`

````python
# Приклад 2: Цикл while з break
print("\nЦикл while з break:")
x = 0
while True:
    print(x)
    x += 1
    if x >= 5:
        break
````

````python
# Приклад 3: Цикл while з continue
print("\nЦикл while з continue:")
y = 0
while y < 5:
    y += 1
    if y == 3:
        continue
    print(y)
````

#### 1.3 Цикл `for`

Цикл `for` використовується для **ітерації по послідовностях** (списки, кортежі, рядки, словники тощо). Кількість ітерацій визначається довжиною послідовності.

**Синтаксис:**
```python
for елемент in послідовність:
    # блок коду, що повторюється
```

````python
# Приклад 4: Простий цикл for з range
print("\nЦикл for:")
for number in range(5):
    print(number)
````

````python
# Приклад 5: Цикл for по списку
print("\nЦикл for по списку:")
fruits = ["apple", "banana", "orange"]
for fruit in fruits:
    print(fruit)
````

````python
# Приклад 6: Цикл for з enumerate (отримання індексу та значення)
print("\nЦикл for з enumerate:")
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")
````

````python
# Приклад 7: Цикл for з break та continue
print("\nЦикл for з break:")
for number in range(10):
    if number == 6:
        break
    print(number)

print("\nЦикл for з continue:")
for number in range(5):
    if number == 2:
        continue
    print(number)
````

#### 1.4 Вкладені цикли

Цикл може містити всередині інший цикл. Такі конструкції називаються **вкладеними циклами**. Кількість рівнів вкладеності не обмежена.

````python
# Приклад 8: Вкладені цикли
print("\nВкладені цикли:")
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
````

````python
# Приклад 9: Цикл for з else
print("\nЦикл for з else:")
for number in range(5):
    print(number)
else:
    print("Цикл завершено без break")
````

#### 1.5 Практичні приклади використання циклів

Цикли широко використовуються для різних обчислень та обробки даних.

````python
# Приклад 10: Обчислення факторіалу циклом
print("\nОбчислення факторіалу циклом:")
n = 5
factorial = 1
for i in range(1, n + 1):
    factorial *= i
print(f"Факторіал {n} = {factorial}")

# Приклад 11: Сума чисел циклом
print("\nСума чисел циклом:")
total = 0
for number in range(1, 6):
    total += number
print(f"Сума чисел від 1 до 5 = {total}")
````

````python
# Приклад 12: Обхід словника циклом
print("\nОбхід словника циклом:")
student = {"Ім'я": "Назар", "Вік": 20, "Група": "ІПЗ-23008бск"}
for key, value in student.items():
    print(f"{key}: {value}")
````

### 2. Рекурсія

#### 2.1 Що таке рекурсія?

**Рекурсія** - це техніка програмування, при якій **функція викликає саму себе** для розв'язання задачі. Кожен рекурсивний виклик розв'язує меншу версію тієї ж задачі.

#### 2.2 Структура рекурсивної функції

1. **Базовий випадок (base case)** - умова, при якій рекурсія припиняється
2. **Рекурсивний крок (recursive step)** - виклик тієї ж функції з іншими аргументами

**Увага:** Без базового випадку або з неправильною умовою може виникнути **нескінченна рекурсія**, яка призведе до переповнення стеку (RecursionError).

````python
# Приклад 13: Факторіал через рекурсію
print("\nРекурсія: обчислення факторіалу")

def factorial(n):
    """Рекурсивна функція для обчислення факторіалу"""
    # Базовий випадок
    if n <= 1:
        return 1
    # Рекурсивний крок
    return n * factorial(n - 1)

result = factorial(5)
print(f"Факторіал 5 = {result}")
````

````python
# Приклад 14: Числа Фібоначчі через рекурсію
print("\nРекурсія: числа Фібоначчі")

def fibonacci(n):
    """Рекурсивна функція для обчислення n-го числа Фібоначчі"""
    # Базові випадки
    if n <= 1:
        return n
    # Рекурсивний крок
    return fibonacci(n - 1) + fibonacci(n - 2)

for i in range(8):
    print(f"Фібоначчі({i}) = {fibonacci(i)}")
````

````python
# Приклад 15: Сума списку через рекурсію
print("\nРекурсія: сума списку")

def list_sum(lst):
    """Рекурсивна функція для обчислення суми елементів списку"""
    # Базовий випадок: порожній список
    if len(lst) == 0:
        return 0
    # Рекурсивний крок: перший елемент + сума решти
    return lst[0] + list_sum(lst[1:])

numbers = [1, 2, 3, 4, 5]
print(f"Сума {numbers} = {list_sum(numbers)}")
````

````python
# Приклад 16: Обхід вкладених списків через рекурсію
print("\nРекурсія: обхід вкладених списків")

def flatten_list(lst):
    """Рекурсивна функція для перетворення вкладеного списку в плоский"""
    result = []
    for item in lst:
        # Якщо елемент є списком, рекурсивно обробляємо його
        if isinstance(item, list):
            result.extend(flatten_list(item))
        else:
            result.append(item)
    return result

nested = [1, [2, [3, 4], 5], 6]
print(f"Оригінал: {nested}")
print(f"Плоский список: {flatten_list(nested)}")
````

#### 2.3 Обмеження рекурсії

У Python глибина рекурсії обмежена (зазвичай близько 1000 рівнів). Це обмеження можна змінити, але важливо завжди мати базовий випадок.

````python
# Приклад 17: Рекурсія з обмеженням глибини
print("\nРекурсія з обмеженням глибини")

def print_recursive(depth, max_depth):
    """Функція з обмеженням глибини рекурсії"""
    # Базовий випадок: досягнуто максимальну глибину
    if depth > max_depth:
        return
    
    print(f"Глибина: {depth}")
    # Рекурсивний виклик
    print_recursive(depth + 1, max_depth)

print_recursive(1, 3)
````

### 3. Порівняння циклів та рекурсії

#### 3.1 Переваги та недоліки

| Критерій | Цикли | Рекурсія |
|----------|-------|----------|
| **Продуктивність** | Швидше, менше накладних витрат | Повільніше через виклики функцій |
| **Використання пам'яті** | Менше пам'яті | Більше пам'яті (стек викликів) |
| **Читабельність** | Прості задачі читаються легко | Складні задачі можуть бути зрозумілішими |
| **Обмеження** | Без обмежень ітерацій | Обмежена глибина рекурсії |

#### 3.2 Коли використовувати?

- **Цикли**: коли кількість ітерацій відома або обробка послідовностей
- **Рекурсія**: для задач, що природно розкладаються на підзадачі (дерева, фрактали, пошук у глибину)

````python
# Приклад 18: Порівняння циклу та рекурсії для факторіалу
print("\nПорівняння циклу і рекурсії для факторіалу:")

n = 5

# Розв'язання циклом
fact_loop = 1
for i in range(1, n + 1):
    fact_loop *= i
print(f"Циклом: факторіал {n} = {fact_loop}")

# Розв'язання рекурсією
def fact_recursive(n):
    if n <= 1:
        return 1
    return n * fact_recursive(n - 1)

print(f"Рекурсією: факторіал {n} = {fact_recursive(n)}")
````

### 4. Додаткові приклади рекурсії

````python
# Приклад 19: Рекурсивне обчислення ступеня
print("\nРекурсивне обчислення ступеня")

def power(base, exponent):
    """Рекурсивна функція для обчислення ступеня"""
    if exponent == 0:
        return 1
    return base * power(base, exponent - 1)

print(f"2^3 = {power(2, 3)}")
print(f"5^2 = {power(5, 2)}")
````

````python
# Приклад 20: Пошук максимального елемента
print("\nПошук максимального елемента")

numbers_list = [3, 7, 2, 9, 1, 5]

# Циклом
max_number = numbers_list[0]
for num in numbers_list:
    if num > max_number:
        max_number = num
print(f"Циклом: максимальне число = {max_number}")

# Рекурсією
def find_max(lst):
    """Рекурсивна функція для пошуку максимального елемента"""
    if len(lst) == 1:
        return lst[0]
    max_rest = find_max(lst[1:])
    return lst[0] if lst[0] > max_rest else max_rest

print(f"Рекурсією: максимальне число = {find_max(numbers_list)}")
````
